```python
# ÌòÑÎåÄ ÏûêÎèôÏûêÎäî Îß§Ìï¥ Ï∂úÏãúÎêòÎäî Ïã†Ìòï ÏûêÎèôÏ∞®Ïùò ÏóêÎÑàÏßÄ ÏÜåÎπÑÌö®Ïú® Îì±Í∏âÏùÑ 1Îì±Í∏âÏúºÎ°ú Ïú†ÏßÄÌïòÍ≥†
# ÏûàÎã§. 22ÎÖÑ Í∞úÎ∞úÎêú Ïã†Ìòï Î™®Îç∏Ïù¥ ÌïúÍµ≠ ÏûêÎèôÏ∞® ÌèâÍ∞ÄÏõêÏóêÏÑú ÏÑ§Ï†ïÌïú ÏóêÎÑàÏßÄ ÏÜåÎπÑ Ìö®Ïú®Îì±Í∏â 1
# Îì±Í∏âÏùÑ Î∞õÏùÑ Ïàò ÏûàÏùÑÏßÄ Í≤ÄÏ†ïÌïòÎ†§ÌïúÎã§. ÌèâÍ∞ÄÏõêÏóê Îî∞Î•¥Î©¥ 1Îì±Í∏âÏùò Í∏∞Ï§ÄÏùÄ ÌèâÍ∑† Î≥µÌï© ÏóêÎÑàÏßÄ
# ÏÜåÎπÑÌö®Ïú®Ïù¥ 16.0 Ïù¥ÏÉÅÏù∏ Í≤ΩÏö∞ Î∂ÄÏó¨ÌïúÎã§Í≥† ÌïúÎã§.

# Îã§ÏùåÏùÄ Ïã†Ìòï ÏûêÎèôÏ∞® 15ÎåÄÏùò Î≥µÌï© ÏóêÎÑàÏßÄÏÜåÎπÑÌö®Ïú® Ï∏°Ï†ïÌïú Í≤∞Í≥ºÏù¥Îã§.
# 15.078, 15.752, 15.549, 15.56, 16.098, 13.277, 15.462, 16.116, 15.214, 16.93,
# 14.118, 14.927, 15.382, 16.709, 16.804

# ÌëúÎ≥∏Ïóê ÏùòÌïòÏó¨ ÌåêÎã®Ìï¥Î≥ºÎïå, ÌòÑÎåÄÏûêÎèôÏ∞®Ïùò Ïã†Ìòï Î™®Îç∏ÏùÄ ÏóêÎÑàÏßÄ Ìö®Ïú® 1Îì±Í∏âÏúºÎ°ú ÌåêÎã®Ìï†
# Ïàò ÏûàÏùÑÏßÄ ÌåêÎã®Ìï¥Î≥¥ÏãúÏò§. (Ïú†ÏùòÏàòÏ§Ä 1%Î°ú ÏÑ§Ï†ï)









import numpy as np
import pandas as pd
import scipy.stats as sp
import matplotlib.pyplot as plt

# 1. ùë° Í≤ÄÏ†ïÏùÑ ÏúÑÌïú Í∞ÄÏ†ïÏ≤¥ÌÅ¨Î•º ÏßÑÌñâÌïòÏÑ∏Ïöî.
#
# ÌëúÎ≥∏ ÌèâÍ∑† , ÌëúÎ≥∏ ÌëúÏ§ÄÌé∏Ï∞® , ÌëúÎ≥∏ Í∞úÏàò
# 15.531733333333333, 0.9802946690129646, 15
# 
x = [ 15.078, 15.752, 15.549, 15.56, 16.098, 13.277, 15.462, 16.116, 15.214, 16.93, 14.118, 14.927, 15.382, 16.709, 16.804]
x_bar = np.mean(x)
s = np.std(x, ddof=1)
n = len(x)
x_bar, s, n      # (15.531733333333333, 0.9802946690129646, 15)


#
# t Í≤ÄÏ†ïÏùÑ ÏúÑÌïú t Î∂ÑÌè¨ Ï†ÅÏö©ÏùÑ ÏúÑÌï¥ÏÑúÎäî Ìï¥Îãπ ÌëúÎ≥∏Ïù¥ Ï†ïÍ∑úÎ∂ÑÌè¨ÏóêÏÑú Ï∂îÏ∂úÎêòÏóàÏùåÏóê ÎåÄÌïú Í∞ÄÏ†ï ÌôïÏù∏Ïù¥ ÌïÑÏöî
# QQ Plot Í≥º shapiro wilk Î°ú Ï†ïÍ∑úÏÑ± Í≤ÄÌÜ†
#
# Q-Q Plot ÌÜµÌïú ÏãúÍ∞ÅÏ†Å ÌôïÏù∏ 
percentile = sp.rankdata(x) / (n + 1)
theoretical_z = sp.norm.ppf(percentile, loc = 0, scale = 1)  # ÌëúÏ§Ä Ï†ïÍ∑úÎ∂ÑÌëúÏÉÅÏóêÏÑú Ï≤¥ÌÅ¨
real_z = (x - x_bar) / s                                     # ‚òÖ ÌëúÏ§ÄÌôî ÊôÇ ÌèâÍ∑†Í≥º ÌëúÏ§Ä Ìé∏Ï∞®Îäî Î¨¥ÏóáÏùÑ ÏÇ¨Ïö©ÌïòÍ≥† Í∑∏ Ïù¥Ïú†Îäî Î¨¥ÏóáÏù∏Í∞Ä?
plt.scatter(theoretical_z, real_z)
plt.plot([-2, -2], [2,2], "g--")
plt.show()
# Ïã§Ï†ú Î∂ÑÏÇ∞ ÎåÄÎπÑ Ïù¥Î°†Ï†Å Î∂ÑÏÇ∞Ïùò ÎπÑÏú®Í∞íÏùÄ shapiro-wilk ÌôïÏù∏ 0.945 
sp.shapiro(x)   # statistic=0.9448550343513489, pvalue=0.447342187166214

# 2. Í≤ÄÏ†ïÏùÑ ÏúÑÌïú Í∞ÄÏÑ§ÏùÑ Î™ÖÌôïÌïòÍ≤å ÏÑúÏà†ÌïòÏãúÏò§.
#
# Null Hypothesis : Î™® ÌèâÍ∑†ÏùÄ 16 Ïù¥ÏÉÅ , Alternative : Î™® ÌèâÍ∑†ÏùÄ 15 ÎØ∏Îßå
# 

# 3. Í≤ÄÏ†ïÌÜµÍ≥ÑÎüâ Í≥ÑÏÇ∞ÌïòÏãúÏò§.
# 4. p‚ÄëvalueÏùÑ Íµ¨ÌïòÏÑ∏Ïöî.
#
# Ïú†ÏùòÏàòÏ§Ä 1% (0.01) Î≥¥Îã§ ÌÅ∞ p-value Î•º Í∞ñÍ≥† ÏûàÏñ¥ Null Hypothesis Í∏∞Í∞ÅÌïòÏßÄ Î™ªÌï®
#
H0 = 16
t_statistics = (x_bar - H0) / (s / np.sqrt(n))    # -1.8500447456376756 ‚Üí t Î∂ÑÌè¨ ÌèâÍ∑†Î≥¥Îã§ ÏôºÏ™ΩÏóê ÏûàÏùå
p_value = sp.t.cdf(x = t_statistics, df = n - 1)  # 0.042762417664207845 ‚Üí 4.27% ‚Üí Ïú†ÏùòÏàòÏ§ÄÎ≥¥Îã§ ÌÅº

#
# Ïù¥Îäî Ïú†ÏùòÏàòÏ§ÄÏóê Ìï¥ÎãπÌïòÎäî Í∏∞Í∞ÅÏπòÎäî 14.867 Ïù¥Í≥† Ï£ºÏñ¥ÏßÑ ÌëúÎ≥∏Ïùò ÌèâÍ∑†ÏùÄ Ïù¥Í≤ÉÎ≥¥Îã§ ÌÅ∞ 15.53 Ïù¥ÎØÄÎ°ú Í∏∞Í∞ÅÌïòÏßÄ Î™ªÌïòÎäî Í≤ÉÏùÑ ÏùòÎØ∏Ìï®
#
t_cr = sp.t.ppf(q=0.01, df = n - 1)               # -2.6244940675602315
cr = H0 + t_cr * (s/np.sqrt(n))                   # 15.335711694757268  

# 5. Ï£ºÏñ¥ÏßÑ Ïú†ÏùòÏàòÏ§ÄÏóê ÎåÄÏùëÌïòÎäî Í∏∞Í∞ÅÏó≠ÏùÄ ÏñºÎßàÏù∏Í∞ÄÏöî?
# 6. ÌòÑÎåÄÏûêÎèôÏ∞®Ïùò Ïã†Ìòï Î™®Îç∏Ïùò ÌèâÍ∑† Î≥µÌï© ÏóêÎÑàÏßÄ ÏÜåÎπÑÌö®Ïú®Ïóê ÎåÄÌïòÏó¨ 95% Ïã†Î¢∞Íµ¨Í∞ÑÏùÑ Íµ¨Ìï¥Î≥¥ÏÑ∏Ïöî.
#
# Ï∞∏Í≥†Î°ú Í∏∞Í∞ÅÏó≠Ïù¥ 14.877 Î≥¥Îã§ ÏûëÏùÄ Î≤îÏúÑÏù¥ÎØÄÎ°ú Ïã†Î¢∞Íµ¨Í∞ÑÏùÄ 14.877 Î≥¥Îã§ ÌÅ∞ Î≤îÏúÑ 
#

# 7. ÌèâÍ∑† Î≥µÌï©ÏóêÎÑàÏßÄ ÏÜåÎπÑÌö®Ïú®Ïù¥ 15Ïù¥ÎùºÎ©¥, ÏúÑÏùò Í≤ÄÏ†ïÏùò Í≤ÄÏ†ïÎ†•ÏùÑ Íµ¨ÌïòÏÑ∏Ïöî. (Îã®, Î™®Î∂ÑÌè¨Ïùò ÌëúÏ§ÄÌé∏Ï∞®Î•º 1ÎùºÍ≥† Í∞ÄÏ†ïÌïúÎã§.)
#
# ÎòêÌïú Î™®ÌèâÍ∑†Ïù¥ 15 ÏùºÎïåÏùò Í≤ÄÏ†ïÎ†•ÏùÄ 97.02% ÏûÑ ‚òÖ (Îã®, Î™®Î∂ÑÌè¨Ïùò ÌëúÏ§ÄÌé∏Ï∞®Î•º 1ÎùºÍ≥† Í∞ÄÏ†ïÌïúÎã§.) Ïù¥Í±∏ Í∞ÄÏ†ïÌïòÎäî Ïù¥Ïú†Í∞Ä Î¨¥ÏóáÏù∏Í∞Ä? Z Î∂ÑÌè¨Î°úÌï¥ÎèÑ Í¥úÏ∞ÆÎã§Îäî Ïù¥ÏïºÍ∏∞ÏûÑ!!!!
#
Ha = 15
t_statisticsa = (cr - Ha) / (s / np.sqrt(n))              # 1.3263418072354864
power = sp.t.cdf(x = t_statisticsa, df = n -1)            # 0.8970210494237376

# ÏúÑÏùò ÌòÑÎåÄ ÏûêÎèôÏ∞® Î¨∏Ï†úÏóêÏÑú Ï£ºÏñ¥ÏßÑ ÌëúÎ≥∏Îì§ÏùÑ ÏÇ¨Ïö©Ìï¥ÏÑú Ïã†Ìòï ÏûêÎèôÏ∞®Ïùò ÌèâÍ∑† ÏóêÎÑàÏßÄ ÏÜåÎπÑ
# Ìö®Ïú®Ïù¥ 16Ïù∏ÏßÄ ÏïÑÎãåÏßÄÎ•º Ïú†ÏùòÏàòÏ§Ä 5%ÌïòÏóêÏÑú Í≤ÄÏ†ïÏùÑ ÏßÑÌñâÌïòÎ†§Í≥† ÌïúÎã§. (Îã®, Î™®Î∂ÑÌè¨Ïùò
# ÌëúÏ§ÄÌé∏Ï∞®Î•º 2ÎùºÍ≥† Í∞ÄÏ†ïÌïúÎã§.)

# 1. Ïã†Ìòï ÏûêÎèôÏ∞®Ïùò ÌèâÍ∑† ÏóêÎÑàÏßÄ ÏÜåÎπÑÌö®Ïú®Ïù¥ 15ÏùºÎïå, Ï£ºÏñ¥ÏßÑ Í≤ÄÏ†ïÏùò Í≤ÄÏ†ïÎ†•ÏùÑ Íµ¨ÌïòÏÑ∏Ïöî.

z_cr1 = sp.norm.ppf(q=0.025)      # -1.9599639845400545
z_cr2 = sp.norm.ppf(q=0.975)      # +1.9599639845400545

x_bar_cr1 = 16 + z_cr1 * 2 / np.sqrt(n)    # 14.987878950494672
x_bar_cr2 = 16 + z_cr2 * 2 / np.sqrt(n)    # 17.01212104950533

z1 = (x_bar_cr1 - 15) / (2 / np.sqrt(n))   # -0.023472311436346135
z2 = (x_bar_cr2 - 15) / (2 / np.sqrt(n))   # 3.8964556576437634

a = sp.norm.cdf(z1)                  # 0.4906367623355539
b = 1 - sp.norm.cdf(z2)              # 4.8805334414980805e-05
power = a + b                        # 0.4906855676699689

# 2. Ïã†Ìòï ÏûêÎèôÏ∞®Ïùò ÌèâÍ∑† ÏóêÎÑàÏßÄ ÏÜåÎπÑÌö®Ïú®Ïù¥ 15ÏùºÎïå, Ï£ºÏñ¥ÏßÑ Í≤ÄÏ†ïÏùò Í≤ÄÏ†ïÎ†•ÏùÑ 80%Î°ú
# ÎßåÏ°±ÏãúÌÇ§ÎèÑÎ°ù ÌïòÎäî ÌëúÎ≥∏Í∞ØÏàòÎ•º Íµ¨Ìï¥Î≥¥ÏÑ∏Ïöî.
N = 0
POWER = 0
for n in range(1,100):
    z_cr1 = sp.norm.ppf(q=0.025)      
    z_cr2 = sp.norm.ppf(q=0.975)      
    x_bar_cr1 = 16 + z_cr1 * 2 / np.sqrt(n)    
    x_bar_cr2 = 16 + z_cr2 * 2 / np.sqrt(n)    
    z1 = (x_bar_cr1 - 15) / (2 / np.sqrt(n))   
    z2 = (x_bar_cr2 - 15) / (2 / np.sqrt(n))   
    a = sp.norm.cdf(z1)                  
    b = 1 - sp.norm.cdf(z2)              
    power = a + b                        
    if (power >= 0.8):
        N = n
        POWER = power
        break

N, POWER
```

```python
import pandas as pd
import numpy as np
import scipy.stats as sp
from scipy.stats import t
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# 1. ùë° Í≤ÄÏ†ïÏùÑ ÏúÑÌïú Í∞ÄÏ†ïÏ≤¥ÌÅ¨Î•º ÏßÑÌñâÌïòÏÑ∏Ïöî.

# t Í≤ÄÏ†ïÏùÄ ÌëúÎ≥∏Ïù¥ Ï†ïÍ∑úÎ∂ÑÌè¨ÏóêÏÑú Ï∂îÏ∂úÎê®ÏùÑ Í∞ÄÏ†ïÌïòÎØÄÎ°ú Q-Q Plot ÏùÑ ÌÜµÌï¥ Ï†ïÍ∑úÏÑ± Í≤ÄÌÜ†
x = np.array([15.078, 15.752, 15.549, 15.56, 16.098, 13.277, 15.462, 16.116, 15.214, 16.93, 14.118, 14.927, 15.382, 16.709, 16.804])

# Ïù¥Î°†Ï†Å Îç∞Ïù¥ÌÑ∞ : Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞Í∞Ä Ï†ïÍ∑úÎ∂ÑÌè¨ÏÉÅÏóê ÏûàÎã§Í≥† Í∞ÄÏ†ïÌïòÎ©¥
percentile = sp.rankdata(x) / (len(x) + 1)
theoretical = norm.ppf(percentile, loc = np.mean(x), scale = np.std(x))  # ‚òÖ Ïó¨Í∏∞ÏÑú ÌëúÎ≥∏Ïùò ÌÜµÍ≥ÑÎüâÏùÑ Í∞ÄÏßÄÍ≥† Ïù¥Î°†Ï†Å ÌôïÎ•†Í∞íÏùÑ Íµ¨ÌïòÎäî Í∑ºÍ±∞Îäî?

# Q-Q Plot
plt.plot([13,17],[13,17], "g--")
plt.scatter(x, theoretical)
plt.show()

# Shapiro-Wilk 
sp.shapiro(x) #tatistic=0.9448550343513489, pvalue=0.447342187166214
```

```python
# 2. Í≤ÄÏ†ïÏùÑ ÏúÑÌïú Í∞ÄÏÑ§ÏùÑ Î™ÖÌôïÌïòÍ≤å ÏÑúÏà†ÌïòÏãúÏò§.

# Í∑ÄÎ¨¥Í∞ÄÏÑ§ : 22ÎÖÑ Ïã†Ìòï ÏûêÎèôÏ∞®Ïùò 'Î≥µÌï© ÏóêÎÑàÏßÄ Ìö®Ïú®' ÌèâÍ∑†ÏùÄ 16.0 Ïù¥ÏÉÅÏù¥Îã§
# ÎåÄÎ¶ΩÍ∞ÄÏÑ§ : 22ÎÖÑ Ïã†Ìòï ÏûêÎèôÏ∞®Ïùò 'Î≥µÌï© ÏóêÎÑàÏßÄ Ìö®Ïú®' ÌèâÍ∑†ÏùÄ 16.0 ÎØ∏ÎßåÏù¥Îã§
```

```python
# 3. Í≤ÄÏ†ïÌÜµÍ≥ÑÎüâ Í≥ÑÏÇ∞ÌïòÏãúÏò§.

# P(X_BAR >= 16)
# [(X_BAR - Î™®ÌèâÍ∑†) / (ÌëúÎ≥∏ ÌëúÏ§ÄÌé∏Ï∞® / ‚àö ÌëúÎ≥∏Í∞úÏàò)] >= [(16 - Î™®ÌèâÍ∑†) / (ÌëúÎ≥∏ ÌëúÏ§ÄÌé∏Ï∞® / ‚àö ÌëúÎ≥∏Í∞úÏàò)]

x_bar = np.mean(x) # ÌëúÎ≥∏ÌèâÍ∑† : 15.531733333333333
s = np.std(x, ddof=1) # ‚òÖ delta degree of freedom ÏÑ§Ï†ï Í∞ÄÎä•!!!
n = len(x)

t_statistics = (x_bar - 16) / (s / n ** 0.5) # -1.8500447456376756
```

```python
# 4. p‚ÄëvalueÏùÑ Íµ¨ÌïòÏÑ∏Ïöî.

t.cdf(t_statistics, df = len(x) - 1) # p-value : 0.042762417664207845

# Í≤ÄÏ†ï ÌÜµÍ≥ÑÎüâ Î≥¥Îã§ Îçî Í∑πÎã®ÏπòÍ∞Ä Í¥ÄÏ∞∞Îê† ÌôïÎ•†ÏùÄ 4.2% ÏàòÏ§Ä
# Ï£ºÏñ¥ÏßÑ Ïú†ÏùòÏàòÏ§ÄÏùÄ 1% Î≥¥Îã§ ÌÅ¨Í∏∞ ÎïåÎ¨∏Ïóê Í∑ÄÎ¨¥Í∞ÄÏÑ§ Í∏∞Í∞ÅÌïòÏßÄ Î™ªÌï® 
```

```python
# 5. Ï£ºÏñ¥ÏßÑ Ïú†ÏùòÏàòÏ§ÄÏóê ÎåÄÏùëÌïòÎäî Í∏∞Í∞ÅÏó≠ÏùÄ ÏñºÎßàÏù∏Í∞ÄÏöî?

t_sig_level = t.ppf(q=0.01, df = len(x) -1)
x_sig_level = t_sig_level * (np.std(x,ddof=1)/np.sqrt(len(x))) + 16

x_sig_level # 15.335711694757268

# ‚Üí ÌëúÎ≥∏ ÌèâÍ∑†Ïùò Í∏∞Í∞ÅÏó≠ÏùÄ ÎßàÏù¥ÎÑàÏä§ Î¨¥ÌïúÎåÄ Î∂ÄÌÑ∞ 15.336 ÏÇ¨Ïù¥Ïùò Î≤îÏúÑÏù¥Îã§
```

```python
# 6. ÌòÑÎåÄÏûêÎèôÏ∞®Ïùò Ïã†Ìòï Î™®Îç∏Ïùò ÌèâÍ∑† Î≥µÌï© ÏóêÎÑàÏßÄ ÏÜåÎπÑÌö®Ïú®Ïóê ÎåÄÌïòÏó¨ 95% Ïã†Î¢∞Íµ¨Í∞ÑÏùÑ
# Íµ¨Ìï¥Î≥¥ÏÑ∏Ïöî.

t_conf_interval = t.ppf(q=0.05, df = len(x) - 1) # t_conf_level

x_conf_interval = np.mean(x) - t_conf_level * (np.std(x, ddof=1)/np.sqrt(len(x))) # 15.977540294592114
t_conf_interval, x_conf_interval

# ‚Üí Î™®ÏßëÎã®ÏùÄ 95% ÌôïÎ•†Î°ú 15.9975 Ïù¥ÏÉÅÏùò Í∞íÏùÑ Í∞ñÎäîÎã§. Ïã†Î¢∞Íµ¨Í∞Ñ (15.9975, INF) ??? ‚òÖ ÎãµÏùÄ (-INF, 15.9975) ÏùºÍπå?
```

```python
# 7. ÌèâÍ∑† Î≥µÌï©ÏóêÎÑàÏßÄ ÏÜåÎπÑÌö®Ïú®Ïù¥ 15Ïù¥ÎùºÎ©¥, ÏúÑÏùò Í≤ÄÏ†ïÏùò Í≤ÄÏ†ïÎ†•ÏùÑ Íµ¨ÌïòÏÑ∏Ïöî. 
# (Îã®, Î™®Î∂ÑÌè¨Ïùò ÌëúÏ§ÄÌé∏Ï∞®Î•º 1ÎùºÍ≥† Í∞ÄÏ†ïÌïúÎã§.)

# P (X <= Í∏∞Í∞ÅÏπò | Î™®ÌèâÍ∑† = 15) Î•º ÎåÄÎ¶ΩÍ∞ÄÏÑ§Ïù¥ Ï∞∏ÏùºÎïåÎ•º Í∞ÄÏ†ïÌïòÍ≥† Íµ¨Ìï®

ta = (x_sig_level - 15) / ( np.std(x, ddof=1) / np.sqrt(len(x)) ) # 1.3263418072354864

t.cdf(ta,df = len(x)-1) # 0.8970210494237376

# 89.7% Í≤ÄÏ†ïÎ†• 
```
