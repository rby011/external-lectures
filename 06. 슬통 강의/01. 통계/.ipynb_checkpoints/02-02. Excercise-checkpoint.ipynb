```python
# 현대 자동자는 매해 출시되는 신형 자동차의 에너지 소비효율 등급을 1등급으로 유지하고
# 있다. 22년 개발된 신형 모델이 한국 자동차 평가원에서 설정한 에너지 소비 효율등급 1
# 등급을 받을 수 있을지 검정하려한다. 평가원에 따르면 1등급의 기준은 평균 복합 에너지
# 소비효율이 16.0 이상인 경우 부여한다고 한다.

# 다음은 신형 자동차 15대의 복합 에너지소비효율 측정한 결과이다.
# 15.078, 15.752, 15.549, 15.56, 16.098, 13.277, 15.462, 16.116, 15.214, 16.93,
# 14.118, 14.927, 15.382, 16.709, 16.804

# 표본에 의하여 판단해볼때, 현대자동차의 신형 모델은 에너지 효율 1등급으로 판단할
# 수 있을지 판단해보시오. (유의수준 1%로 설정)









import numpy as np
import pandas as pd
import scipy.stats as sp
import matplotlib.pyplot as plt

# 1. 𝑡 검정을 위한 가정체크를 진행하세요.
#
# 표본 평균 , 표본 표준편차 , 표본 개수
# 15.531733333333333, 0.9802946690129646, 15
# 
x = [ 15.078, 15.752, 15.549, 15.56, 16.098, 13.277, 15.462, 16.116, 15.214, 16.93, 14.118, 14.927, 15.382, 16.709, 16.804]
x_bar = np.mean(x)
s = np.std(x, ddof=1)
n = len(x)
x_bar, s, n      # (15.531733333333333, 0.9802946690129646, 15)


#
# t 검정을 위한 t 분포 적용을 위해서는 해당 표본이 정규분포에서 추출되었음에 대한 가정 확인이 필요
# QQ Plot 과 shapiro wilk 로 정규성 검토
#
# Q-Q Plot 통한 시각적 확인 
percentile = sp.rankdata(x) / (n + 1)
theoretical_z = sp.norm.ppf(percentile, loc = 0, scale = 1)  # 표준 정규분표상에서 체크
real_z = (x - x_bar) / s                                     # ★ 표준화 時 평균과 표준 편차는 무엇을 사용하고 그 이유는 무엇인가?
plt.scatter(theoretical_z, real_z)
plt.plot([-2, -2], [2,2], "g--")
plt.show()
# 실제 분산 대비 이론적 분산의 비율값은 shapiro-wilk 확인 0.945 
sp.shapiro(x)   # statistic=0.9448550343513489, pvalue=0.447342187166214

# 2. 검정을 위한 가설을 명확하게 서술하시오.
#
# Null Hypothesis : 모 평균은 16 이상 , Alternative : 모 평균은 15 미만
# 

# 3. 검정통계량 계산하시오.
# 4. p‑value을 구하세요.
#
# 유의수준 1% (0.01) 보다 큰 p-value 를 갖고 있어 Null Hypothesis 기각하지 못함
#
H0 = 16
t_statistics = (x_bar - H0) / (s / np.sqrt(n))    # -1.8500447456376756 → t 분포 평균보다 왼쪽에 있음
p_value = sp.t.cdf(x = t_statistics, df = n - 1)  # 0.042762417664207845 → 4.27% → 유의수준보다 큼

#
# 이는 유의수준에 해당하는 기각치는 14.867 이고 주어진 표본의 평균은 이것보다 큰 15.53 이므로 기각하지 못하는 것을 의미함
#
t_cr = sp.t.ppf(q=0.01, df = n - 1)               # -2.6244940675602315
cr = H0 + t_cr * (s/np.sqrt(n))                   # 15.335711694757268  

# 5. 주어진 유의수준에 대응하는 기각역은 얼마인가요?
# 6. 현대자동차의 신형 모델의 평균 복합 에너지 소비효율에 대하여 95% 신뢰구간을 구해보세요.
#
# 참고로 기각역이 14.877 보다 작은 범위이므로 신뢰구간은 14.877 보다 큰 범위 
#

# 7. 평균 복합에너지 소비효율이 15이라면, 위의 검정의 검정력을 구하세요. (단, 모분포의 표준편차를 1라고 가정한다.)
#
# 또한 모평균이 15 일때의 검정력은 97.02% 임 ★ (단, 모분포의 표준편차를 1라고 가정한다.) 이걸 가정하는 이유가 무엇인가? Z 분포로해도 괜찮다는 이야기임!!!!
#
Ha = 15
t_statisticsa = (cr - Ha) / (s / np.sqrt(n))              # 1.3263418072354864
power = sp.t.cdf(x = t_statisticsa, df = n -1)            # 0.8970210494237376

# 위의 현대 자동차 문제에서 주어진 표본들을 사용해서 신형 자동차의 평균 에너지 소비
# 효율이 16인지 아닌지를 유의수준 5%하에서 검정을 진행하려고 한다. (단, 모분포의
# 표준편차를 2라고 가정한다.)

# 1. 신형 자동차의 평균 에너지 소비효율이 15일때, 주어진 검정의 검정력을 구하세요.

z_cr1 = sp.norm.ppf(q=0.025)      # -1.9599639845400545
z_cr2 = sp.norm.ppf(q=0.975)      # +1.9599639845400545

x_bar_cr1 = 16 + z_cr1 * 2 / np.sqrt(n)    # 14.987878950494672
x_bar_cr2 = 16 + z_cr2 * 2 / np.sqrt(n)    # 17.01212104950533

z1 = (x_bar_cr1 - 15) / (2 / np.sqrt(n))   # -0.023472311436346135
z2 = (x_bar_cr2 - 15) / (2 / np.sqrt(n))   # 3.8964556576437634

a = sp.norm.cdf(z1)                  # 0.4906367623355539
b = 1 - sp.norm.cdf(z2)              # 4.8805334414980805e-05
power = a + b                        # 0.4906855676699689

# 2. 신형 자동차의 평균 에너지 소비효율이 15일때, 주어진 검정의 검정력을 80%로
# 만족시키도록 하는 표본갯수를 구해보세요.
N = 0
POWER = 0
for n in range(1,100):
    z_cr1 = sp.norm.ppf(q=0.025)      
    z_cr2 = sp.norm.ppf(q=0.975)      
    x_bar_cr1 = 16 + z_cr1 * 2 / np.sqrt(n)    
    x_bar_cr2 = 16 + z_cr2 * 2 / np.sqrt(n)    
    z1 = (x_bar_cr1 - 15) / (2 / np.sqrt(n))   
    z2 = (x_bar_cr2 - 15) / (2 / np.sqrt(n))   
    a = sp.norm.cdf(z1)                  
    b = 1 - sp.norm.cdf(z2)              
    power = a + b                        
    if (power >= 0.8):
        N = n
        POWER = power
        break

N, POWER
```

```python
import pandas as pd
import numpy as np
import scipy.stats as sp
from scipy.stats import t
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# 1. 𝑡 검정을 위한 가정체크를 진행하세요.

# t 검정은 표본이 정규분포에서 추출됨을 가정하므로 Q-Q Plot 을 통해 정규성 검토
x = np.array([15.078, 15.752, 15.549, 15.56, 16.098, 13.277, 15.462, 16.116, 15.214, 16.93, 14.118, 14.927, 15.382, 16.709, 16.804])

# 이론적 데이터 : 주어진 데이터가 정규분포상에 있다고 가정하면
percentile = sp.rankdata(x) / (len(x) + 1)
theoretical = norm.ppf(percentile, loc = np.mean(x), scale = np.std(x))  # ★ 여기서 표본의 통계량을 가지고 이론적 확률값을 구하는 근거는?

# Q-Q Plot
plt.plot([13,17],[13,17], "g--")
plt.scatter(x, theoretical)
plt.show()

# Shapiro-Wilk 
sp.shapiro(x) #tatistic=0.9448550343513489, pvalue=0.447342187166214
```

```python
# 2. 검정을 위한 가설을 명확하게 서술하시오.

# 귀무가설 : 22년 신형 자동차의 '복합 에너지 효율' 평균은 16.0 이상이다
# 대립가설 : 22년 신형 자동차의 '복합 에너지 효율' 평균은 16.0 미만이다
```

```python
# 3. 검정통계량 계산하시오.

# P(X_BAR >= 16)
# [(X_BAR - 모평균) / (표본 표준편차 / √ 표본개수)] >= [(16 - 모평균) / (표본 표준편차 / √ 표본개수)]

x_bar = np.mean(x) # 표본평균 : 15.531733333333333
s = np.std(x, ddof=1) # ★ delta degree of freedom 설정 가능!!!
n = len(x)

t_statistics = (x_bar - 16) / (s / n ** 0.5) # -1.8500447456376756
```

```python
# 4. p‑value을 구하세요.

t.cdf(t_statistics, df = len(x) - 1) # p-value : 0.042762417664207845

# 검정 통계량 보다 더 극단치가 관찰될 확률은 4.2% 수준
# 주어진 유의수준은 1% 보다 크기 때문에 귀무가설 기각하지 못함 
```

```python
# 5. 주어진 유의수준에 대응하는 기각역은 얼마인가요?

t_sig_level = t.ppf(q=0.01, df = len(x) -1)
x_sig_level = t_sig_level * (np.std(x,ddof=1)/np.sqrt(len(x))) + 16

x_sig_level # 15.335711694757268

# → 표본 평균의 기각역은 마이너스 무한대 부터 15.336 사이의 범위이다
```

```python
# 6. 현대자동차의 신형 모델의 평균 복합 에너지 소비효율에 대하여 95% 신뢰구간을
# 구해보세요.

t_conf_interval = t.ppf(q=0.05, df = len(x) - 1) # t_conf_level

x_conf_interval = np.mean(x) - t_conf_level * (np.std(x, ddof=1)/np.sqrt(len(x))) # 15.977540294592114
t_conf_interval, x_conf_interval

# → 모집단은 95% 확률로 15.9975 이상의 값을 갖는다. 신뢰구간 (15.9975, INF) ??? ★ 답은 (-INF, 15.9975) 일까?
```

```python
# 7. 평균 복합에너지 소비효율이 15이라면, 위의 검정의 검정력을 구하세요. 
# (단, 모분포의 표준편차를 1라고 가정한다.)

# P (X <= 기각치 | 모평균 = 15) 를 대립가설이 참일때를 가정하고 구함

ta = (x_sig_level - 15) / ( np.std(x, ddof=1) / np.sqrt(len(x)) ) # 1.3263418072354864

t.cdf(ta,df = len(x)-1) # 0.8970210494237376

# 89.7% 검정력 
```
