```python
#
# ----------------------------------------------------------------
#                   ANOVA (Analysis of Variance)
# ----------------------------------------------------------------
# 1. 개요 : 분산을 활용하여 모평균의 비교
# 2. 방식 
#  (1) One-way ANOVA
#    - Concept : 관찰값(xij)은 xij = μi + εij 와 같고 노이즈(εij)는 εij ~ N(0, σ²) 임 
#
#    - Statistics : F = MSG / MSE ~ F (k - 1)(n - k)     
#      * MSG (Mean Squared Error for Group) :  { ∑ (관찰값 전체 평균과 각 그룹의 평균의 편차 제곱) * (그룹의 크기) } /  { (그룹 개수 - 1) }   
#      * MSE (Mean Squared Error) : ∑ { (그룹의 표본 분산) * (그룹의 크기 - 1) } / { (표본 개수 - 그룹 개수) }
#      * k   (Number of groups)  ,  n   (Number of observations)
#      → MSG 가 크고 MSE 가 작으면 (그룹간 이질성, 그룹내 동질성) 서로 다르다라고 판단  
#
#    - Assumption : εij ~ N(0, σ²)
#      ① 평균이 0 으로 같은지 여부
#      ② 분산이 σ² 으로 같은지 여부
#      ③ 정규성을 갖는지 여부
#
#    - Example : 향수 종류에 따른 고객의 매장 체류 시간의 평균을 비교 
#                * 향수 종류(라벤더, 레몬, 무향) → 범주형 변수 / 체류 시간 → 연속형 변수
#                * 귀무가설 : 라벤더, 레몬, 무향의 평균 체류 시간은 같다
#                * 대립가설 : 라벤더, 레몬, 무향의 평규 체류 시간은 하나라도 다르다
#
#    - Hierachy 
#                                    Y
#                               --------->   statsmodels.stats.anova.anova_lm() with statsmodels.formula.api.ols() 
#                              |             scipy.stats.f_oneway(equal_var = TRUE)
#                              |             (post) statsmodels.stats.multicomp.tukeyhsd()
#            -------> 등분산성 
#            |                 |     N
#          Y |                  --------->   scipy.stats.f_oneway(equal_var = False)
#            |                               (post) scipy.stats.ttest_ind(var_equal = False) * group 간의 평균 차이 정도만
#      정규성  
#            |                       Y
#          N |                  --------->   scipy.stats.kruskal() → median
#            |                 |             (post) wilcoxson rank sum, man whitney U * median 검정해야 하니..
#            -------> 등분산성 
#                              |     N
#                               --------->   max(Sk) < 2 * min(Sk) 조건 내內 scipy.stats.kruskal() → median
#                                            (post) wilcoxson rank sum, man whitney U * median 검정해야 하니..
# 
#  (2) Two-way ANOVA (하단 정리)
# ----------------------------------------------------------------
```

```python
#
# [1] Statistics 으로 개념 잡아보기 
# - F = MSG / MSE ~ F (k - 1) ( n - 1)
#

import pandas as pd
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt

df1 = pd.read_csv('data/anova_example.csv')

k = df1["Odor"].value_counts().sum() # No_odor,  Lavender,  Lemon

MSG = 0
total_mean = df1["Minutes"].mean()
table = df1.groupby("Odor").mean().reset_index()
table.columns = ["odor", "Mean"]
table["n"] = df1.groupby("Odor").count().values
SSG = (table["n"] * (table["Mean"] - total_mean) ** 2).sum()
MSG = SSG / (k - 1)

MSE = 0
table["S2"] = df1.groupby("Odor")["Minutes"].var(ddof = 1) .values
MSE = (( table["n"] - 1 ) * table["S2"] / (table["n"] - 1)).sum()

F = MSG / MSE
dist = sp.chi2(k - 1, df1.shape[0] - k)

x = dist.rvs(size = 10000)
y = dist.pdf(x)
plt.scatter(x, y)

plt.scatter(F, 0)
plt.show()
```

```python
#
# [2] One-way ANOVA - A. Simple Analysis 
#

import pandas as pd
import numpy as np
import scipy.stats as sp

df = pd.read_csv("./data/anova_example.csv")

grps = []
for idx in df['Odor'].value_counts().index:
    grps = grps + [df[df["Odor"] == idx]["Minutes"]]

sp.f_oneway(grps[0], grps[1], grps[2]) # statistic=6.700198443966929, pvalue=0.0030925029998487064

#
# f_oneway 의 문제는 가정검정, 사후 검정을 하기가 번거롭다는 것
#  - 가정검정 : 잔차의 정규성, 등분산성, 등평균
#  - 사후검정 : 어떤 그룹의 평균이 다른 것인지 확인
# 
# → 그래서 Odor 에 따른 Minutes 을 예측하는 회귀모델을 적합하고
#   이 모델이 가지고 있는 Odor 그룹 별 잔차 등의 데이터를 활용해
#   다양한 통계량을 도출할 수 있는 anova_lm 을 이용하는 것이 적절 
```

```python
#
# [3] One-way ANOVA - B. Proper Analysis
#

import numpy as np
import pandas as pd
import scipy.stats as sp
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as smc 
import matplotlib.pyplot as plt
import pingouin as pg
import seaborn as sns

df = pd.read_csv('data/anova_example.csv')

df1 = pd.get_dummies(df['Odor'], prefix='Odor')
df1["Minutes"] = df["Minutes"].copy()

formula = "Minutes ~  " + " + ".join(df1.columns[0:-1])
model = smf.ols(formula, data=df1).fit()

#
# I. 가설 검정 : 잔차의 정규성, 평균이 0, 등분산성 (왜냐면 모분산을 모르니)
#

# a. 잔차 정규성 검토
print("# 잔차의 정규성 검토 ")
pg.qqplot(model.resid)
plt.show()
print(" - 검정 : " , sp.shapiro(model.resid))
print()

# b. 그룹별 잔차 평균이 0 인지
print("# 그룹별 잔차의 평균 검토")

df["Residual"] = model.resid
df["Params"] = 0
df.loc[df['Odor'] == 'Lavender', 'Params'] = model.params["Odor_Lavender"]
df.loc[df['Odor'] == 'Lemon', 'Params'] = model.params["Odor_Lemon"]
df.loc[df['Odor'] == 'No_odor', 'Params'] = model.params["Odor_No_odor"]
plt.scatter(df[df["Odor"] == "Lavender"]["Params"], df[df["Odor"] == "Lavender"]["Residual"])
plt.scatter(df[df["Odor"] == "Lemon"]["Params"], df[df["Odor"] == "Lemon"]["Residual"])
plt.scatter(df[df["Odor"] == "No_odor"]["Params"], df[df["Odor"] == "No_odor"]["Residual"])
plt.plot(model.params[1:],df.groupby(["Odor"])["Residual"].mean())
plt.show()

t1 = sp.ttest_1samp(df.loc[df['Odor'] == 'Lavender', 'Residual'],0)
t2 = sp.ttest_1samp(df.loc[df['Odor'] == 'Lemon', 'Residual'],0)
t3 = sp.ttest_1samp(df.loc[df['Odor'] == 'No_odor', 'Residual'],0)

print("- Lavender T-Test :", t1)
print("- Lemon T-Test :", t2)
print("- No_odor T-Test :", t3)
print()

# c. 그룹간 분산이 같은지

print("# 그룹간의 등분산성")

sns.boxplot(x="Odor", y="Residual", data=df)
plt.show()

l1 = sp.levene(df.loc[df['Odor'] == 'Lavender', 'Residual'], 
                df.loc[df['Odor'] == 'Lemon', 'Residual'],
                df.loc[df['Odor'] == 'No_odor', 'Residual'])
print("- Leven Test :", l1)
print()

#
# II. ANOVA 실행
#
print("# ANOVA 실행")
aov_result = sm.stats.anova_lm(model, typ=2)  # anova table for fitted linear model
print("- ANOVA : \n", aov_result)
print()

#
# III. 사후검정
#
print("# 사후검정 - p-value 가 엄청 작게 나왔으니...")
post = smc.MultiComparison(df["Minutes"], df["Odor"])
p_result = post.tukeyhsd()
print("- Tukey HSD : \n", p_result)
```

```python
#
# [3] 정규성이 만족하지 않는 경우의 Kruskal-Wallis 검정
#

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, levene, kruskal

# 그룹 데이터 생성
group1 = np.random.normal(0, 1, 30)
group2 = np.random.normal(1, 1, 30)
group3 = np.random.normal(2, 1, 30)

# 정규성 검정
stat1, p1 = shapiro(group1)
stat2, p2 = shapiro(group2)
stat3, p3 = shapiro(group3)
print("Shapiro-Wilk 검정 결과(p-value) = ", p1, p2, p3)

# 등분산성 검정
stat, p = levene(group1, group2, group3)
print("Levene 검정 결과(p-value) = ", p)

# 정규성 시각화
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
axs[0].hist(group1, bins=10)
axs[0].set_title("Group 1")
axs[1].hist(group2, bins=10)
axs[1].set_title("Group 2")
axs[2].hist(group3, bins=10)
axs[2].set_title("Group 3")
fig.suptitle("Normality Test", fontsize=16)
plt.show()

# 등분산성 시각화
plt.boxplot([group1, group2, group3], labels=["Group 1", "Group 2", "Group 3"])
plt.title("Equal Variance Test")
plt.show()

# Kruskal-Wallis H 검정
H, p = kruskal(group1, group2, group3)
print("Kruskal-Wallis H 검정 결과(H, p-value) = ", H, p)
```

```python
#
# ---------------------------------------------------------------- 
#                         Two-way ANOVA
# ----------------------------------------------------------------
#
# (1) Concept :  관찰값(xijk)은 xijk = μij + εijk 와 같고 노이즈(εijk)는 εijk ~ N(0, σ²) 임 
# 
# (2) Main Effect, Interaction Effect 
#  - 개념 모델 : Y = μ + A + B + AB + ε   ( A, B = main effect , AB = interaction effect ) 
#  - Two-way ANOVA 검정은 두 개의 범주형 변수(인자)에 따른 연속형 변수(반응 변수)의 차이를 분석
#     . Main Effect : 두 인자가 종속 변수에 미치는 영향
#     . Interaction Effect : 두 인자가 함께 종속 변수에 미치는 영향 
#  - Main Effect, Interaction Effect 는 서로 독립적인 것 (크기나 방향에 서로 영향을 주지 않음)
#    단, Interfaction Effect 가 없다면 Main Effect 의 크기가 더욱 의미를 가질 수 있는 있음
#  
# (3) Two-way ANOVA 의 가정 차이 :  관찰값의 크기가 요인별 그룹의 크기가 동일하다는 가정 有 
```

```python
#
# (3) 그룹 크기에 따른 신뢰성 차이 유有
#  - 동일 분포 서로 크기가 다르게 group 을 추출한 것, 동일하게 group 을 추출한 것 사이에
#  - f-statistics, p-value 가 서로 다름. 동일한 것의 p-value 가 더 작음     
#
import numpy as np
from scipy.stats import f

# 그룹별 크기가 다른 경우와 같은 경우의 데이터 생성
np.random.seed(123)
group1 = np.concatenate([np.random.normal(loc=5, size=10),
                         np.random.normal(loc=10, size=10),
                         np.random.normal(loc=15, size=5)])
group2 = np.concatenate([np.random.normal(loc=10, size=20),
                         np.random.normal(loc=15, size=20)])

# 그룹별 크기가 다른 경우와 같은 경우의 Two-way ANOVA 검정 수행
def two_way_anova(group1, group2):
    n1, n2 = len(group1), len(group2)
    mean1, mean2 = np.mean(group1), np.mean(group2)
    ssw1 = np.sum((group1 - mean1) ** 2)
    ssw2 = np.sum((group2 - mean2) ** 2)
    ssw = ssw1 + ssw2
    ssb1 = n1 * (mean1 - np.mean([mean1, mean2])) ** 2
    ssb2 = n2 * (mean2 - np.mean([mean1, mean2])) ** 2
    ssb = ssb1 + ssb2
    sst = ssw + ssb
    dfw = n1 + n2 - 2
    dfb = 1
    dft = dfw + dfb
    msw = ssw / dfw
    msb = ssb / dfb
    f_stat = msb / msw
    p_value = f.sf(f_stat, dfb, dfw)
    return f_stat, p_value

f_stat1, p_value1 = two_way_anova(group1, group2)
print("그룹별 크기가 다른 경우의 F-value: {:.2f}, p-value: {:.4f}".format(f_stat1, p_value1))

group1_equal_size = np.concatenate([np.random.normal(loc=5, size=10),
                                    np.random.normal(loc=10, size=10),
                                    np.random.normal(loc=15, size=10)])
group2_equal_size = np.concatenate([np.random.normal(loc=10, size=10),
                                    np.random.normal(loc=15, size=10),
                                    np.random.normal(loc=20, size=10)])

f_stat2, p_value2 = two_way_anova(group1_equal_size, group2_equal_size)
print("그룹별 크기가 같은 경우의 F-value: {:.2f}, p-value: {:.4f}".format(f_stat2, p_value2))
```
