```python
#
# [1] 확률밀도함수 , i.e., 특정 확률변수 X 의 P(X) 의 값 구하기
#  ① scipy.stats 가 지원하는 확률밀도함수들 : https://docs.scipy.org/doc/scipy/reference/stats.html
#    [연속확률분포]
#      - t         : A Student's t continuous random variable.
#      - norm      : A normal continuous random variable.
#      - chi2      : A chi-squared continuous random variable.
#      - expon     : An exponential continuous random variable.
#      - f         : An F continuous random variable.
#    [이산확률분포]
#      - bernoulli : A Bernoulli discrete random variable.
#      - poisson   : A Poisson discrete random variable.
# ② 확률분포 별 지원 함수 (commonly)
#    [정규분포 사례]
#      - pdf(x, loc=0, scale=1) : Probability density function. ※ x 지점의 값
#      - cdf(x, loc=0, scale=1) : Cumulative distribution function. ※ 확률변수 X 의 최소값부터 주어진 x까지의 누적
#  

import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.stats import t
import matplotlib.pyplot as plt

# 정규분포 함수에 확률변수 X 를 넣어 값을 얻어옴
x = np.arange(-5, 10, 0.1)
y = norm.pdf(x, loc=3, scale=2) # loc : 평균 , scale : 표준편차
plt.plot(x, y, color='k')
```

```python
# t 분포 함수에 확률변수 X 를 넣어 값을 얻어옴
t_x = np.arange(-5, 10, 0.1)
t_y = t.pdf(x, len(x) - 1)
plt.plot(t_x, t_y, color='c')
```

```python
#
# [2] 분포에서 표본 추출하기
#  ※ 아래와 같은 분포 등에서 표본을 추출할 수 있음 
#  =================== =========================================================
#  Univariate
#  distributions
#  -------------------- ---------------------------------------------------------
#  chisquare            :math:`\chi^2` distribution.
#  exponential          Exponential distribution.
#  normal               Normal / Gaussian distribution.
#  poisson              Poisson distribution.
#  uniform              Uniform distribution.
#  ==================== =========================================================
#  =================== =========================================================
#  Standard
#  distributions
#  -------------------- ---------------------------------------------------------
#  standard_normal      Standard normal distribution.
#  standard_t           Standard Student's t-distribution.
#  standard_exponential Standard exponential distribution.
# ==================== =========================================================

# Draw random samples from a normal (Gaussian) distribution.
# - loc : mean of distribution
# - scale : standard deviation (spread or width) of the distribution
# - size : output shape
np.random.normal(loc = 3, scale=2, size=5)
# array([3.46256024, 4.7511354 , 5.86536763, 3.81915913, 3.96311295])

# Draw samples from a standard Normal distribution (mean=0, stdev=1).
np.random.standard_normal(size=5)
# array([-0.8773984 ,  0.97765336, -0.12797896, -0.89940742,  2.74373318])

# Draw samples from a standard Student's t distribution with `df` degrees of freedom.
# 정리하자면, 정규분포 모집단에서 추출한 표본평균을 표본 표준편차로 표준화하면 
# 자유도 n -1 인 분포를 하게 t 분포를 얻게됩니다.
np.random.standard_t(df = 15, size = 5)
# array([ 1.9884543 ,  2.02350468,  0.34342752,  0.25878146, -0.46545739])
```

```python
#
# [3] 분포에서 누적확률 구하기 e.g) 4.5 보다 큰 숫자가 나올 확률 구하기
# 

# P( -∞ <= X <= 4.5)
# 평균이 3 이고 표준편차가 2인 정규분포상에서
# -∞ ~ 4.5 까지의 누적 확률
p1 = norm.cdf(4.5, loc =3, scale = 2)

# P(X>=4.5)
# 4.5 보다 큰 숫자가 나올 확률
p2 = 1 - p1

p1, p2
```

```python
#
# [3] 분포에서 누적확률이 q 인 확률변수 X 의 값 구하기
#

# cent point function (inverse of `cdf`) at q of the given RV.
# - q: lower tail probability
# - loc , scale
x = norm.ppf(q = p1, loc = 3, scale = 2)


# 평균이 3 이고 표준편차 2 인 정규분포에서 상위 10% 에 해당하는 X 는?
x2 = norm.ppf(q = 0.9 , loc = 3, scale = 2)
```

```python
#
# [4] 68, 95, 99.7 의 정규분포의 Rule of Thumb
#

# 평균이 3 이고 표준편차가 2 인 정규분포에서 10,000 개 샘플 추출
x = np.random.normal(loc = 3, scale = 2, size = 10000)

# 1 시그마 범위(평균 ± 1 * 표준편차)내에 포함되는 표본 데이터의 비율
len(x[(x > 3-2) & (x < 3 + 2) ]) / 10000          # 68.9%

# 2 시그마 범위(평균 ± 2 * 표준편차)내에 포함되는 표본 데이터의 비율
len(x[(x > (3-2*2)) & (x < (3+2*2))]) / 10000     # 95.82%

# 2 시그마 범위(평균 ± 3 * 표준편차)내에 포함되는 표본 데이터의 비율
len(x[(x > (3-3*2)) & (x < (3+3*2))]) / 10000     # 99.84%
```

```python
#
# [5] 표준화된 정규분포는 Z ~ N (0, 1^2) 을 따른다
#

# (결론) 정규분포에서 뽑은 샘플을 표준화 한 것은 표준 정규분포 결과와 유사함
#       - A, B 의 그래프가 거의 동일하므로 표준화가 잘 동작함

# A. 정규분포에서 추출한 샘플을 표준화 과정을 거친것
#
# N(3,2^2) 인 정규분포에서 10,000개의 샘플에 대해서 표준화 후
# kernel density estimation 한 결과
x1 = np.random.normal(loc = 3, scale = 2, size = 10000)
z1 = (x1 - 3) / 2
df1 = pd.DataFrame(z1)
df1.plot.density()

# B. 임의 샘플 추출한 것을 표준정규분포함수에 대입한 결과를 구한 것
#
# 표준편차 2 를 고려해 -3 , 3 까지 대략적으로 임의 샘플링을 한 
# 후에 표준 정규분포 함수를 통해 확률값을 구한 결과
x2 = np.arange(start = -3, stop = 3, step = 6/10000)
y2 = norm.pdf(x2, loc = 0, scale = 1)
plt.plot(x2, y2)
```

```python
#
# [6] 대수의 법칙
#

# 추출한 표본이 클수록 표본의 평균은 모수의 평균을 따름
# - 여기서의 모수의 평균은 5 , 표본의 평균은 그래프상의 중심
for n in range(10, 1001, 100):
    # 정규분포에서만 적용되는 것은 아님. 여기서는 평균 고정하려고 정규분포에서 추출함
    x = np.random.normal(loc = 5, scale = 2, size = n)  
    df = pd.DataFrame(x)
    df.plot.density()

# 수치적으로 뽑아봄
for n in range(10, 1001, 100):
    x = np.random.normal(loc = 5, scale = 2, size = n)  
    print(n, np.mean(x))
```

```python
# 
# [7] 중심극한정리
#  - ( X_bar - μ ) * n^0,5  ~  N (0, σ^2)
#    ← X_bar - μ 는 0 에 가까워지므로 n^0.5 를 곱해서 분포를 보니 정규분포를 따름
#
# * Simulation : https://onlinestatbook.com/stat_sim/sampling_dist/index.html
#
for n in range(10, 100001, 10000):
    x = np.arange(start = -5, stop = 5, step = 10/n)   # 어떠한 분포에서도 마찬가지이다!!! 여기서는 완전 랜덤 샘플
    df = pd.DataFrame(x)
    print(df.describe())
    
df.plot.density(bw_method=0.5)
```
