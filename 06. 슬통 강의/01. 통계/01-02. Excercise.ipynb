### 문제 4) 어떤 시험의 표준점수는 평균이 30 , 표준편차가 5 인 정규분포를 따른다고 한다.

### 
### 1) 표준점수의 밀도함수를 그려라
### 2) 수험생을 임의로 1명 선택하여 점수를 조회하였을 때 45 점 보다 높은 점수를 받았을 확률은?
### 3) 수험생 A 는 상위 10% 에 해당하는 점수를 얻었다고 한다면, 수험생A의 점수는 얼마인가?
### 4) 수험생이 16명이었다고 할 때, 16 명의 평균 점수를 내었을 때, 이 평균값이 따라는 분포의 확률밀도 함수를 1번 그래와 겹쳐서 그려라
### 5) 이 때 평균 점수가 38 점보다 높게 나올 확률은?

```python
# 1) 정규분포의 밀도함수

import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 3 표준편차 정도의 범위면 99% 정도 데이터 표현 가능한데 좀더 넓혀보면
x = np.arange(start = 30 - 3.5 * 5, stop = 30 + 3.5 * 5, step = 0.1)
y = norm.pdf(x, loc = 30, scale = 5)
plt.plot(x, y, "r")
plt.show()
```

```python
# 2) 45 점 보다 높은 점수를 받았을 확률 : P(X>45)

p = 1 - norm.cdf(45, loc = 30, scale = 5) # cdf : P(-∞ <= X <=45) 
p * 100  # 0.135% 확률
```

```python
# 3) 수험생 A 는 상위 10% 에 해당하는 점수를 얻었다고 한다면, 수험생A의 점수는 얼마인가?

norm.ppf(q = 0.9, loc = 30, scale = 5)  # P(X) 가 0.9 인 X 값은
```

```python
# 4) 수험생이 16명이었다고 할 때, 16 명의 평균 점수를 내었을 때, 이 평균값이 따르는 분포의 확률밀도 함수를 1번 그래와 겹쳐서 그려라
# - 모집단 통계량 : 평균 30, 표준편차 5

x = np.arange(start = 30 - 3.5 * 5, stop = 30 + 3.5 * 5, step = 0.1)
y = norm.pdf(x, 30, 5)
# P ( √n ( x_bar - μ) ) ~ N (0 , σ^2) → P (x_bar) ~ N (μ, (σ/√n)^2 )
x_bar = np.arange(start = 30 - 3.5 * 5, stop = 30 + 3.5 * 5, step = 0.1)
# y_bar = norm.pdf(x_bar, 30, (5/16**0.5)**2) ← Wrong 인수는 분산이 아닌 표준편차
y_bar = norm.pdf(x_bar, 30, (5/16**0.5)) 

plt.plot(x,y, "r")
plt.plot(x_bar, y_bar, "g")
plt.show() # 평균의 분포는 표본의 분포에 비해 꼬리가 얇음. 그럴수 밖에 없을듯
```

### 4) 자유도가 4인 카이제곱분포를 따르는 확률변수에서 나올 수 있는 값 중 상위 5% 에 해당하는 값은 얼마인가?

### 6) 평균이 3, 표준편차가 2 인 정규분포를 따르는 확률변수에서 크기가 20인 표본 x1, ... , x20 을 뽑은 후 표본분산을 계산한 것을 s₁² 이라고 하자
###    - 같은 방법으로 500 개의 s₁², s₂² , ... 을 발생시켜 보시오
###    - 발생한 500 개의  s² 들 각각에 4.75 를 곱하고 그것들의 히스토그램을 그려보시오 (히스토그램을 그릴 때 probability = TRUE 사용)
###    - 위에서 그린 히스토그램에 자유도가 19인 카이제곱분포 확률밀도 함수를 겹쳐 그려보시오

```python
# 1) 자유도가 4인 카이제곱분포의 확률밀도 함수

from scipy.stats import chi2 

x = np.arange(0, 30, 0.01)
y = chi2.pdf(x, df = 4)

plt.plot(x, y)
plt.show()

# 2) P ( 3 ≤ X ≤ 5) 인 확률은
chi2.cdf(5, df = 4) - chi2.cdf(3, df = 4)  # 0.27052790518742903
```

```python
# 3) 자유도가 4인 카이제곱분포에서 크기가 1,000 인 표본을 뽑은 후 히스토그램을 그리시오

from scipy.stats import chi2
import matplotlib.pyplot as plt

x = chi2.rvs(df=4, loc=0, scale=1, size=1000)
plt.hist(x) # x 의 빈도를 세는 것이 histogram 임!!
plt.show()

x2 = np.random.chisquare(df=4, size = 1000)
plt.hist(x2) # x 의 빈도를 세는 것이 histogram 임!!
plt.show()

# → 분포에서 random sampling 하게되면 2) 의 확률분포와 엇비슷하게 x 의 빈도가 출현!!
```

```python
# 4) 자유도가 4인 카이제곱분포를 따르는 확률변수에서 나올 수 있는 값 중 상위 5% 에 해당하는 값

chi2.ppf(q = 0.95, df = 4) # 9.487729036781154
```

```python
# 5) 3번에서 뽑인 표본값들 중 상위 5% 에 위치한 표본 값은 얼마인가

df = pd.DataFrame()
df["x"] = x
df.x.quantile(q=0.95, interpolation='nearest') # 9.348802141598352
df.x.quantile(q=0.95, interpolation='nearest') # 9.35065584527733
```

```python
# 6)
# 평균이 3, 표준편차가 2 인 정규분포를 따르는 확률변수에서 크기가 20인 표본 x1, ... , x20 을 뽑은 후 표본분산을 계산한 것을 s₁² 이라고 하자

# - 같은 방법으로 500 개의 s₁², s₂² , ... 을 발생시켜 보시오
x_vars = []
for i in range(0,500,1):
    x = norm.rvs(loc = 3, scale=2, size=500)
    x_var = (sum((np.mean(x) - x) ** 2) / (len(x) - 1)) 
    x_vars = x_vars + [x_var]
x_vars = np.array(x_vars)

# - 발생한 500 개의 s² 들 각각에 4.75 를 곱하고 그것들의 히스토그램을 그려보시오 (히스토그램을 그릴 때 probability = TRUE 사용)¶
plt.hist(x_vars * 4.75,density=True)

# - 위에서 그린 히스토그램에 자유도가 19인 카이제곱분포 확률밀도 함수를 겹쳐 그려보시오
x_c = chi2.rvs(df=19, size=1000)
y_c = chi2.pdf(x = x_c, df=19)
plt.plot(x_c,y_c)

plt.show()
```
