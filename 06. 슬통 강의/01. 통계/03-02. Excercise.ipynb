```python
# Q1
# 성인 남성 100명의 표본 키의 표본 평균 키가 168cm 이고 모집단의 표준편차가 2.5 cm 인 것을 고려하여
# 성인 남성의 모집 평균 키에 대한 95% 신뢰 구간을 설정하라
# 상단측(오른쪽이 뚫린) 으로 구할 것

import scipy.stats as sp
import numpy as np

# 표준 정규 분포상에서 임계치는 -1.64 로 평균의 왼편에 통계량이 존재함 
168 - sp.norm.ppf(0.05) * (2.5 / np.sqrt(100))

# 신뢰구간은 (168.41, INF) 
```

```python
# Q2
# 표본 평균 중량이 10kg 이고 표본의 표준편차가 0.2kg 인 밀가루 100봉지의 표본을 고려하여
# 특정 회사에서 생산한 밀가루 봉지의 단측 평균 신뢰구간 90% 를 구하시오
# 하단측으로 구할 것

# 표본의 개수가 100 개로 표본의 편차를 모편차로 보기에 충분히 큼
# 따라서 z 분포를 이용해 신뢰구간을 구함
z = sp.norm.ppf(q=0.9)    # 1.2815515655446004

# 90% 신뢰 구간으로 모평균이 위치할 수 있는 최대 임계치
u = 10 + (0.2 / np.sqrt(100)) * z    # 10.025631031310892

# 신뢰 구간은 (-INF, 10.0256)
```

```python
# Q3
# 한 사탕회사에서 사탕봉지의 평균 무게가 목표 무게인 1파운드와 같은지 여부를 확인하려한다.
# 이 회사는 25개의 사탕봉지의 표본을 수집했고 그것들의 평균 문제가는 0.95 파운드, 표준편차는 0.2 파운드이다.
# t 검정의 결론은 무엇이며 사탕봉지의 실제 모집단 평균 무게에 대한 신뢰구간을 구하라
# 양측으로 구할것

# 표본의 개수가 충분하지 않아 t 분포를 이용해 추정
# 또한 표본이 정규성을 가정하고 t 검정을 수행하면
n = 25
x_bar = 0.95
s = 0.2
u = 1

# 귀무가설 : 모평균 u = 1 , 대립가설 u 는 1 이 아님
# 귀무가설이 참이라고 가정한 분포상에서의 t 통계량과 대응하는 p-value 
t_statistics = (x_bar - u) / (s / np.sqrt(n))           # -1.250000000000001
p_value = sp.t.cdf(t_statistics , df = n-1) * 2         # 0.22335147816561993
# 유의확률이 22% 이상으로 일반적 유의수준 5% 보다 크기 때문에 귀무가설을 받아들임. 
# → 모평균은 1 임

# 유의수준 5% , 신뢰도 95% 전제하에서의 신뢰구간은
t_a = sp.t.ppf(q=0.025, df = n-1)                      # -2.063898561628021
t_b = sp.t.ppf(q=0.975, df = n-1)                      # 2.0638985616280205
u_a = x_bar + t_a * (s / np.sqrt(n))
u_b = x_bar + t_b * (s / np.sqrt(n))

# 모평균은 95% 의 신뢰도를 가지고 아래 구간에 위치함
(u_a,u_b)                                              # (0.8674440575348791, 1.0325559424651207)
```

```python
# Q4
# 새로 제안된 혈압약에 대한 효과 분석을 위하여 무작위로 배정된 두 그룹에 대한 혈압
# 측정 데이터 이다. Treated 그룹의 경우 혈압약을 일정기간 복용하였으며, Control 그
# 룹은 평상시 활동을 그대로 유지하였다. 표 3.4는 두 그룹의 혈압을 측정한 데이터이다.
# 혈압약의 사용이 혈압을 떨어뜨리는 효과가 있는지 유의수준 5%하에서 검정하시오.

# ID Score Group
# 1 3.81 control
# 2 4.47 control
# 3 4.81 control
# 4 4.79 control
# 5 4.25 control
# 6 3.93 treated
# 7 4.26 treated
# 8 3.74 treated
# 9 3.61 treated
# 10 3.93 treated
# 11 4.36 treated
# 12 3.93 treated
# 13 3.89 treated
# 14 4.03 treated
# 15 3.85 treated
# 16 4.06 treated

# • 귀무가설 vs. 대립가설
# • 가정 체크 및 검정 방법 설정
```

```python
import pandas as pd
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
import pingouin as pg

df4 = pd.read_csv("./data/medicine2.csv")
df4
df_control = df4[df4["Group"] == "control"]["Score"]
df_treated = df4[df4["Group"] == "treated"]["Score"]


#
# H0 : 신약의 효과 없어 u_control <= u_treated 
# Ha : 신약의 효과가 있어 u_control > u_treated
# 

# 정규성 검토
# 1) qq plot
print("# 정규성 검토")
plt.subplot(1,2,1)
plt.title("control")
pg.qqplot(df_control,dist="norm")
plt.subplot(1,2,2)
plt.title("treated")
pg.qqplot(df_treated,dist="norm")
plt.show()
# 2) shapiro-wilk 검정 수행
# H0 : 데이터가 정규분포를 따른다
# Ha : 데이터가 정규분포를 따르지 않는다
print("- control : ", sp.shapiro(df_control))
print("- treated : ", sp.shapiro(df_treated))
# → 유의수준 5% 내 귀무가설 기각 불가

# 등분산성 검정
# H0 : 두 그룹의 모집단 분산이 같다
# Ha : 두 그룹의 모집단 분산은 같지 않다
print()
print("# 등분산성 검토 :", sp.levene(df_control, df_treated,center="mean"))
# → 유의수준 5% 내 귀무가설 기각 불가

# 정규성, 등분산성 검토되었음. 독립표본 t 검정 실시함
# H0 : 신약의 효과 없어 u_control <= u_treated 
# Ha : 신약의 효과가 있어 u_control > u_treated
print()
print("# t 검정 : ", sp.ttest_ind(df_treated,df_control, alternative="less"))
# → p-value 가 유의수준 5% 보다 작아 귀무가설을 기각.  
```

```python
# Q4
# 고양의 소변이 식물의 성장을 저해한다는 명제를 확인하기 위하여 무작위로 선정된 두
# 식물 그룹을 가지고 실험을 진행하였다.
# Treated 그룹의 경우 고양이 소변을 체취하여 일정 간격으로 뿌려주었으며, 이외 다른
# 조건은 Treated, Control 그룹 모두 동일하게 유지하였다. 표 3.5는 한달 후 두 그룹의
# 성장 높이 데이터를 기록한 것이다.

# 고양의 소변이 식물의 성장을 저해한다는 것에 대한 검정을 유의수준 5%하에서 수행 해보시오.

# 표 3.5: 식물 성장 측정 데이터
# ID Score Group
# 1 45 control
# 2 87 control
# 3 123 control
# 4 120 control
# 5 70 control
# 6 51 treated
# 7 71 treated
# 8 42 treated
# 9 37 treated
# 10 51 treated
# 11 78 treated
# 12 51 treated
# 13 49 treated
# 14 56 treated
# 15 47 treated
# 16 58 treated

# • 귀무가설 vs. 대립가설
# • 가정 체크 및 검정 방법 설정
```

```python
import pandas as pd
import numpy as np
import scipy.stats as sp
import pingouin as pg
import matplotlib.pyplot as plt

df4 = pd.read_csv("./data/cat.csv")

df_control = df4[df4["Group"] == "control"]["Score"]
df_treated = df4[df4["Group"] == "treated"]["Score"]

control = np.array(df_control)
treated = np.array(df_treated)

# 데이터의 정규성 검토 및 검증 
# - qqplot 상 크게 벗어나는 데이터는 없었고
# - shapiro-wilk 검정시 유의수준보다 유의확률이 커 귀무가설 받아들임
# - 귀무가설 : control, treated 그룹은 정규성을 따른다
# - 대립가설 : control, treated 그룹은 정규성을 따르지 않는 것이 있다
plt.subplot(1,2,1)
plt.title("control")
pg.qqplot(control)
plt.subplot(1,2,2)
plt.title("treated")
pg.qqplot(treated)
plt.show()
print(sp.shapiro(control), sp.shapiro(treated))  
# control : ShapiroResult(statistic=0.9245066046714783, pvalue=0.5594186186790466) 
# treated : ShapiroResult(statistic=0.9170810580253601, pvalue=0.2950733006000519)

# 데이터의 등분산성 검토 
# 정규성을 따르므로 정규분포상에서 검정력이 다소 우수한 F 검정을 수행
# - 귀무가설 : 두 집단의 모분산은 같다
# - 대립가설 : 두 집단의 모분산은 같지 않다
x = np.arange(0,10,0.01)

f = np.var(control, ddof=1) / np.var(treated, ddof=1)         # 7.788141025641026
plt.scatter(x, sp.f.pdf(x, len(control)-1, len(treated)-1))
plt.scatter(f, 0)
plt.show()
a = 1 - sp.f.cdf(f, len(control)-1, len(treated)-1)           # 0.004054135385040181

f_rev = np.var(treated, ddof=1) / np.var(control, ddof=1)     # 0.1284003456932384
plt.scatter(x, sp.f.pdf(x, len(treated)-1, len(control)-1))
plt.scatter(f_rev, 0)
plt.show()
b = sp.f.cdf(f_rev, len(treated)-1, len(control)- 1)          # 0.0040541353850401364

p_value = a + b                                               # 0.008108270770080316

# Q. ★ 그런데 한쪽만 구해서 x 2 를 해도 그 값이 같은 것으로 보이는데. 이거도 맞는 것일까?
# f_statistics = np.var(control, ddof=1) / np.var(treated,ddof=1)                    # 7.788141025641026
# p_value1 = (1 - sp.f.cdf(f_statistics, len(control) - 1, len(treated) -1)) * 2     # 0.008108270770080361
# a = (1 - sp.f.cdf(f_statistics, len(control) - 1, len(treated) -1))
# b = (sp.f.cdf(1/f_statistics, len(control) - 1, len(treated) -1))
# p_value2 = a +b
# p_value1, p_value2

# f 테스트 어떻게 하는지 모르니 levene 검정 실행하면 p-value 가 유의수준 5% 보다 작아 두 집단의 분산이 같다고 볼 수 없다
# - 검정 데이터는 정규성을 따르는 것으로 정규성 검정시 결과를 얻었음. 따라서 중앙은 평균으로 설정하여 leven test 수행함
# print(sp.levene(control, treated,center="mean"))       # LeveneResult(statistic=8.872339653570032, pvalue=0.009964317348520352)


# 두 집단은 정규성을 만족하나 등분산성은 만족하지 않는 상태이다.
# 따라서 Welch two sample t 검증을 수행한다
# - 귀무가설 : 고양이 소변은 식물 성장을 저해하지 않음. u_control <= u_treated
# - 대립가설 : 고양이 소변은 식물 성장을 저해함 .      u_control > u_treated
# 
# 검정 수행 결과 유의확률이 0.003 (3%) 로 유의수준 5% 보다 작아 귀무가설을 기각하고 대립가설을 받아 들임
# 즉. 고양이 소변은 식물 성장을 저해한다고 볼 수 있고 그 차이는 주어진 유의수준 내에서 통계적으로 유의미하다
# sp.ttest_ind(control, treated,equal_var=False,alternative="greater")  # Ttest_indResult(statistic=2.306873049653553, pvalue=0.03766591739314407)
```
